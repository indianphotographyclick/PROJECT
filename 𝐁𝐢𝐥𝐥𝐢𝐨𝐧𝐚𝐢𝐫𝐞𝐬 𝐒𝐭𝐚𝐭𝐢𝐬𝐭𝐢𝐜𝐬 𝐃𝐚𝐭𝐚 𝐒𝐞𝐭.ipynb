### **Filtering**

#### **Filter all billionaires under 40 years old**

```python
All_Billionaires_Under_40_Years_Old = data[(data['Age'] < 40)]
All_Billionaires_Under_40_Years_Old.head()
```

#### **Filter all billionaires who are both female and self-made**

```python
Female_and_Self_Made = data[(data['Gender'] == 'F')&
                            (data['Self Made'] == True)]
Female_and_Self_Made.head()
```

#### **Filter all billionaires in the Technology industry**

```python
Technology_Industry_In_All_Billionaires = data[(data['Industries'] == 'Technology')]
Technology_Industry_In_All_Billionaires.head()
```

#### **Filter all billionaires with net worth over $50 billion**

```python
All_Billionaire_Fifty_Billion = data[(data['finalWorth'] > 50000)]
All_Billionaire_Fifty_Billion.head()
```

#### **Filter all billionaires from India who are self-made**

```python
All_billionaires_India = data[(data['Country'] == 'India')&
                              (data['Self Made'] == True)]
All_billionaires_India.head()
```

###  **Descriptive Statistics**

####  **What is the average net worth of a billionaire?**

```python
Average_Net_Worth = data['finalWorth'].mean()
f"Average Net Worth of a Billionaire: ${Average_Net_Worth:.2f} million"
```

#### **What is the median net worth of billionaires, and how does it compare to the average?**

```python
Mean_Net_Worth = data['finalWorth'].mean()
Median_Net_Worth = data['finalWorth'].median()
print(f"Mean Net Worth : ${Mean_Net_Worth:.2f} million")
print(f"Median Net Worth : ${Median_Net_Worth:.2f} million")
```

#### **What is the standard deviation of net worth?**

```python
Standard_Deviation = data['finalWorth'].std()
f"Standard Deviation of Net Worth: ${Standard_Deviation:.2f} million"
```

#### **What is the range of ages among billionaires?**

```python
Among_Billionaires_Min_Age = data['Age'].min()
Among_Billionaires_Max_Age = data['Age'].max()
Among_Billionaires_Age_Range = Among_Billionaires_Max_Age - Among_Billionaires_Min_Age

print(f"Youngest Billionaire Age: {Among_Billionaires_Min_Age}")
print(f"Oldest Billionaire Age: {Among_Billionaires_Max_Age}")
print(f"Age Range: {Among_Billionaires_Age_Range} years")
```

#### **Which country has the highest average net worth per billionaire?**

```python
Highest_Average = data.groupby('Country')['finalWorth'].mean().sort_values(ascending=False)
Top_Country = Highest_Average.idxmax()
Top_Avg = Highest_Average.max()

print(f"Country with Highest Average Net Worth: {Top_Country}")
print(f"Average Net Worth: ${Top_Avg:.2f} million")
```

#### **What percentage of billionaires are self-made vs. inherited?**

```python
Self_Made_Counts = data['Self Made'].value_counts(normalize=True) * 100
print("Percentage of Billionaires by Wealth Origin:")
print(Self_Made_Counts.round(2))
```

#### **What is the average age of billionaires by industry?**

```python
Billionaires_Industry = data.groupby('Industries')['finalWorth'].mean().sort_values(ascending=False)
print("Average Age of Billionaires by Industry (Top 10):")
print(Billionaires_Industry.head(10).round(1))
```

#### **What are the top 5 cities by average billionaire net worth?**

```python
Average_Billionaire_Net_Worth_By_City = data.groupby('City')['finalWorth'].mean().sort_values(ascending=False)
print("Top 5 Cities By Average Billionaire Net Worth:")
print(Average_Billionaire_Net_Worth_By_City.head(5).round(2))
```

#### **What is the interquartile range (IQR) of net worth?**

```python
Q1 = data['finalWorth'].quantile(0.25)
Q2 = data['finalWorth'].quantile(0.75)
IQR = Q2 - Q1
print(f"Q1 (25th percentile): ${Q1:.2f} million")
print(f"Q3 (75th percentile): ${Q2:.2f} million")
print(f"IQR (Interquartile Range): ${IQR:.2f} million")
```

#### **How many billionaires fall above the 75th percentile of net worth?**

```python
Q = data['finalWorth'].quantile(0.75)
Top_Quartile_Billionaires = data[data['finalWorth'] > Q]
Count_Top_25_Percent = Top_Quartile_Billionaires.shape[0]
f"Number of Billionaires Above the 75th Percentile: {Count_Top_25_Percent}"
```

### **Visualization** 

#### **Trend of billionaire count by country (Top 10 countries)**

```python
data.columns = data.columns.str.strip()
Top_10_Countries = data['Country'].value_counts().head(10).index
Filtered_Data = data[data['Country'].isin(Top_10_Countries)]
plt.figure(figsize=(12, 6))
sns.countplot(data=Filtered_Data, y='Country', hue='Gender', order=Top_10_Countries, palette='deep')
plt.title('Top 10 Countries by Number of Billionaires (Split by Gender)', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### **Distribution of Billionaire Ages**

```python
data.columns = data.columns.str.strip()
plt.figure(figsize=(10,6))
sns.histplot(data['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution Of Billionaire Ages', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Number of Billionaires')
plt.tight_layout()
plt.show()
```

#### **Age vs Net Worth**

```python
data.columns = data.columns.str.strip()
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='finalWorth', hue='Self Made', palette='Set1', alpha=0.7)
plt.title('Age vs Net Worth (Colored by Self-Made Status)', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Net Worth (Millions USD)')
plt.legend(title='Self Made')
plt.tight_layout()
plt.show()
```

####  **Gender Distribution by Country**

```python
data.columns = data.columns.str.strip()
Top_Countries = data['Country'].value_counts().head(10).index
Filtered_Data = data[data['Country'].isin(Top_Countries)]
Gender_Country_Counts = pd.crosstab(Filtered_Data['Country'], Filtered_Data['Gender'])
Gender_Country_Counts.plot(kind='barh', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Gender Distribution of Billionaires in Top 10 Countries', fontsize=14)
plt.xlabel('Number of Billionaires')
plt.ylabel('Country')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### **Billionaire Count by Country and Industry**

```python
data.columns = data.columns.str.strip()
Top_Countries = data['Country'].value_counts().head(10).index
Top_Industries = data['category'].value_counts().head(10).index
Filtered = data[data['Country'].isin(Top_Countries) & data['category'].isin(Top_Industries)]
Heatmap_Data = pd.crosstab(Filtered['Country'], Filtered['category'])
plt.figure(figsize=(12, 6))
sns.heatmap(Heatmap_Data, annot=True, fmt='d', cmap='Reds')
plt.title('Heatmap: Number of Billionaires by Country and Industry', fontsize=14)
plt.xlabel('Industry')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

#### **Self-Made vs Inherited Wealth**

```python
data.columns = data.columns.str.strip()
Self_Made_Counts = data['Self Made'].value_counts()
plt.figure(figsize=(6, 6))
colors = ['#FA8072', '#87CEEB']
plt.pie(Self_Made_Counts, labels=Self_Made_Counts.index, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'width':0.4})
plt.title('Self-Made vs Inherited Billionaires', fontsize=14)
plt.tight_layout()
plt.show()
```

#### **Number of Billionaires by Age Group**

```python
data.columns = data.columns.str.strip()
bins = [20, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['20‚Äì29', '30‚Äì39', '40‚Äì49', '50‚Äì59', '60‚Äì69', '70‚Äì79', '80‚Äì89', '90+']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
Age_Group_Counts = data['Age Group'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x=Age_Group_Counts.index, y=Age_Group_Counts.values, marker='o', color='darkorange')
plt.title('Number of Billionaires by Age Group', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Number of Billionaires')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### **Top 10 Cities with Most Billionaires**

```python
Top_Cities = data['City'].value_counts().head(10).index
City_Data = data[data['City'].isin(Top_Cities)]
plt.figure(figsize=(12, 6))
sns.countplot(data=City_Data, y='City', hue='Gender', order=Top_Cities, palette='magma')
plt.title('Top 10 Cities with Most Billionaires (Split by Gender)')
plt.xlabel('Number of Billionaires')
plt.ylabel('City')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### **Correlation Between Numerical Features**

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='Reds')
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```

#### **Gender Distribution of Billionaires**
